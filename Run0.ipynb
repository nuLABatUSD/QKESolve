```python
import SOLVING_QKE_FUNCTION as solve
import numpy as np
import derivatives as der
```

```python
N=200
Emax= 10
Eval= np.linspace(Emax/N, Emax, N)
ym0 = np.zeros((N,4))
for i in range(len(ym0)):
    p0 = 1/(np.exp(Eval[i])+1)
    ym0[i,:] = [p0,0,0,1]
ym0_bar= np.zeros((N,4))
for i in range(len(ym0_bar)):
    p0= 1/(np.exp(Eval[i])+1)
    ym0_bar[i,:]= [p0,0,0,1]
```

```python
%%time
T = [0.12, 0.25, 0.44, 1, 2, 4]

foldername = "DataRun0"
basename = "Run"

y0 = der.array_maker(ym0)
for i in range(len(T)):
    %time solve.solve_QKE(T[i], y0, False, False, foldername, basename + '0{}'.format(i), incl_collisions=False)
    %time solve.solve_QKE(T[i], y0, False, False, foldername, basename + '1{}'.format(i), incl_collisions=True)
    %time solve.solve_QKE(T[i], y0, True, False, foldername, basename + '2{}'.format(i), incl_collisions=False)
    %time solve.solve_QKE(T[i], y0, True, False, foldername, basename + '3{}'.format(i), incl_collisions=True)
```
