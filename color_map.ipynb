```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.load("mu01Data/poster_data/plots/2000keV.npz", allow_pickle=True)
settings = data['settings'].item()
raw = data['raw'].item()
nu3D = raw['nu3D']
nubar3D = raw['nubar3D']
```

```python
T = settings['T']
N = settings['N'] + 1
E = np.linspace(0,settings['eps_max'], N) * T
time = raw['time']
```

```python
def dnue_dE(E, P0, Pz):
    return E**2 / (2 * np.pi**2) * 1/2 * P0 * (1+Pz)
def dnumu_dE(E, P0, Pz):
    return E**2 / (2 * np.pi**2) * 1/2 * P0 * (1-Pz)
```

```python
def create_c(function, data):
    c = np.zeros((len(data), len(E[1:])))
    for j in range(len(E[1:])):
        initial = function(E[j+1], data[0,j,0], data[0,j,3])
        for i in range(len(data)):
            c[i,j] = function(E[j+1], data[i,j,0], data[i,j,3]) / initial
    return c.T
```

```python
def one_number_density(t, ddE, data):
    y = np.zeros(len(E[1:]))
    for i in range(len(y)):
        y[i] = ddE(E[i+1], data[t,i,0], data[t,i,3])
    
    return 4 * np.pi**2 / (3 * 1.20206 * T**3) * np.trapz(y,E[1:])
```

```python
def all_number_densities(ddE, data):
    n = np.zeros(len(time))
    for i in range(len(time)):
        n[i] = one_number_density(i, ddE, data)
    return n
```

```python
def plot(function, data, title):
    fig = plt.figure(figsize=[4,3])
    color = plt.axes([0.1, 0, 1.00, 1.11])
    dens = plt.axes([0.1, 1.11, 1.00, 0.5])

    x = time
    y = all_number_densities(function, data)
    dens.plot(x,y)
    dens.set_ylabel("number density/thermal distribution n_nu / n_{nu, th}")
    
    c = create_c(function, data)
    color.imshow(c[:,1:],origin='lower',vmin=0.9,vmax=1.1,aspect='auto',extent=(time[0], time[len(time)-1], E[0], E[len(E)-1]))
    color.set_ylabel("Energy")
    
    dens.xaxis.set_major_formatter(plt.NullFormatter())
    color.set_xlabel("Time")
    plt.title(title)
```

```python
x = time
y1 = all_number_densities(dnue_dE, nu3D)

y2 = all_number_densities(dnue_dE, nubar3D)

c1 = create_c(dnue_dE, nu3D)
c2 = create_c(dnue_dE, nubar3D)

f, axs = plt.subplots(nrows=3,ncols=1,sharex='col', sharey='row', height_ratios=[2,3,3], figsize=(6,6))
plt.subplots_adjust(wspace=0.2, hspace=0)

axs[0].plot(x,y1, label=r'$\nu_e$', color="green")
axs[0].plot(x,y2, label=r'$\bar\nu_e$', color='#430086', linestyle='dashed')
axs[0].legend()
axs[0].set_ylabel(r"$n_{\nu} / n_{\nu, th}$")
axs[1].imshow(c1[1:,:],origin='lower',aspect='auto',vmin=.99, vmax=1.01,extent=(time[0], time[-1], E[0], E[len(E)-1]))
axs[1].set_ylabel(r"$\nu_e ~ energy~ [MeV]$")
im = axs[2].imshow(c2[1:,:],origin='lower',aspect='auto',vmin=.99, vmax=1.01,extent=(time[0], time[-1], E[0], E[len(E)-1]))
axs[2].set_xlabel("Time [ms]")
axs[2].set_ylabel(r"$\bar\nu_e ~ energy ~  [MeV]$")
axs[1].set_yticks(np.arange(2*T,10*T,2*T))
axs[2].set_yticks(np.arange(2*T,10*T,2*T))
axs[2].set_xticks(np.arange(0,1.9*10**(19),.25*10**(19)), labels=[0,.25,.50,.75,1.00,1.25,1.50,1.75])
f.colorbar(im,ax=axs[:],fraction=.02, anchor=(0,.33))
plt.show()
```
